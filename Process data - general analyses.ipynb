```python
import pandas as pd
import numpy as np
```

### colors/values appearence analysis

read Excel files by quadrant

```python
blackbox1_top_left_quadrant = pd.read_csv("blackbox1_top_left_quadrant.csv")
blackbox1_top_right_quadrant = pd.read_csv("blackbox1_top_right_quadrant.csv")
blackbox1_bottom_left_quadrant = pd.read_csv("blackbox1_bottom_left_quadrant.csv")
blackbox1_bottom_right_quadrant = pd.read_csv("blackbox1_bottom_right_quadrant.csv")
```

find missing value by step

```python
l = [1,2,3,4,5,6,7,8,9,0]                #a list contain all values
```

```python
blackbox1_top_left_quadrant.head(5)
```

```python
row = []
missing_value = []
blackbox1_top_left_quadrant["missing values"] = np.nan*4000
for row_index in range(4000):
    row = blackbox1_top_left_quadrant.iloc[row_index, 1:].tolist() #extract a row
    row = np.unique(row)     #get unique values from current row
    #print(row)
    missing_value = list(set(l)-set(row))    #find missing colors/values
    blackbox1_top_left_quadrant.loc[row_index,"missing values"] = ','.join(str(missing_value))
    #print(','.join(missing_value))
```
